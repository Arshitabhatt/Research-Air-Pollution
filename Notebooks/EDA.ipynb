```python
import pandas as pd

series = pd.read_csv("datasetforpaperjan.csv")
df1 = pd.DataFrame(series, columns =['Year','PM10', 'Mortaility', 'Population', 'Vehicles on Road']) 
df2 = pd.DataFrame(series, columns =['Year','SO2', 'Mortaility', 'Population', 'Vehicles on Road']) 
df3 = pd.DataFrame(series, columns =['Year','NO2', 'Mortaility', 'Population', 'Vehicles on Road']) 
df1 = df1.set_index('Year')
df2 = df2.set_index('Year')
df3 = df3.set_index('Year')
df1.info()
df2.info()
df3.info()
```

```python
print(df1.corr(method='pearson'))
print(df1.corr())
print(df3.corr())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(7,7))
sns.heatmap(df1.corr(), annot=True, cmap="YlGnBu").set_title("Correlation of PM10 with various factors")
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.heatmap(df2.corr(), annot=True,cmap="Greens").set_title("Correlation of SO2 with various factors")
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.heatmap(df3.corr(), annot=True).set_title("Correlation of NO2 with various factors")
plt.show()
```
